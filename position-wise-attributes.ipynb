```python
# Imports necessary
import os
import pandas as pd
import numpy as np
import requests
import json
```

```python
# This cell makes the call to fpl api and receives the data in 
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
response = requests.get(url)
response = json.loads(response.content)
```

```python
# Creating dataframes of necessary keys
elements_df = pd.DataFrame(response['elements'])
elements_types_df = pd.DataFrame(response['element_types'])
teams_df = pd.DataFrame(response['teams'])
```

```python
# Replacing position and team from element_types_df
elements_df['position'] = elements_df.element_type.map(elements_types_df.set_index('id').singular_name)
elements_df['team_'] = elements_df.team.map(teams_df.set_index('id').name)
elements_df['value'] = elements_df.value_season.astype(float)
elements_df = elements_df.loc[elements_df.value > 0]
```

```python
common_attributes = elements_df[['second_name','position','bonus','bps','dreamteam_count','influence','minutes',
                                 'now_cost','own_goals','penalties_missed','points_per_game',
                                'red_cards','selected_by_percent','event_points','team_',
                                'total_points','transfers_in_event','transfers_out_event','form',
                                'yellow_cards']]
attack_attributes = elements_df[['second_name','assists','creativity','creativity_rank',
                                 'goals_scored','ict_index','ict_index_rank']]
defence_attributes = elements_df[['second_name','clean_sheets','goals_conceded']]
goalkeeper_attributes = elements_df[['second_name','penalties_saved','saves']]
```

```python
# Combine different dataframes
attacker_attributes = pd.merge(common_attributes,attacking_attributes,on='second_name')
attacker_attributes.head()
```
